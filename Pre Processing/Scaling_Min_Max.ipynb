```
from sklearn.preprocessing import MinMaxScaler

data = features_train.copy() #copying data frame
```

```
del data['genre_x']
del data['genre_y']
"""del data['Author_Name']
del data['Book_Name']"""
del data['genre']
```

```
data.head()
```

```
normalizer = MinMaxScaler()
data_scaled = normalizer.fit_transform(data.iloc[:,data.columns!='ID'])
data_norm = pd.DataFrame(np.column_stack((data['ID'].values,data_scaled)),columns = data.columns).set_index(data.index)
```

```
data_norm.head()
```

```
data_norm.ID.astype(int) #changed data type of column
```

```
data_norm.head(1)
```

```
#merging data frame
features_train_subset = features_train[['ID', 'genre']] #'Author_Name', 'Book_Name'
features_scaled_train = pd.merge(data_norm, features_train_subset , on='ID')
```
